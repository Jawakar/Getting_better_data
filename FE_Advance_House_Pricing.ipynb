```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
pd.pandas.set_option('display.max_columns', None)
```

```python
HPdf = pd.read_csv("/content/drive/MyDrive/Kaggle/train.csv")
HPdf.head()
```

# Fixing Categorical Missing Values

```python
Cat_miss_features = [f for f in HPdf.columns if HPdf[f].isnull().sum()>1 and HPdf[f].dtype == 'O']
for f in Cat_miss_features:
    print(f'{f}: {np.round(HPdf[f].isnull().mean(),4)} missing values')
```

```python
# replacing missing values(NaN) with "missing" label 
def replace_cat_feature(HPdf, f):
    data = HPdf.copy()
    data[f] = data[f].fillna('Missing')
    return data

HPdf = replace_cat_feature(HPdf, Cat_miss_features)
HPdf.head()
```

```python
HPdf.shape
```

# Fixing Numerical misssing value

```python
Num_miss_features = [f for f in HPdf.columns if HPdf[f].isnull().sum()>1 and HPdf[f].dtype != 'O']
for f in Num_miss_features:
    print(f'{f:} {np.round(HPdf[f].isnull().mean(),4)}')
```

```python
# replacing the numerical misssing values

for f in Num_miss_features:
    # as we have seen that more outliers in EDA, we replace with the median value
    median_value = HPdf[f].median()
    # we are creating 1 if Nan else 0 as seperate column
    HPdf[f+"NaN"] = np.where(HPdf[f].isnull(),1,0)
    HPdf[f].fillna(median_value, inplace=True)
HPdf[Num_miss_features].isnull().sum()
```

```python
HPdf.head()
```

```python
# getting numerical feature and year feature
num_features = [f for f in HPdf.columns if HPdf[f].dtype != 'O']
year_features = [f for f in num_features if "Yr" in f or "Year" in f]
year_features
```

```python
# from EDA part we know that YrSold is not good, so we consider no.of years for other features
for f in year_features:
    if f!='YrSold':
        HPdf[f] = HPdf['YrSold']-HPdf[f]
```

```python
HPdf[year_features].head()
```

```python
# we convert skewed distribution features to log normal distribution
skewed_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']
for f in skewed_features:
    HPdf[f] = np.log(HPdf[f])
```

```python
HPdf.head()
```

# Handeleing rare features and encoding

```python
# removing rare categorical features
categorical_features = [f for f in HPdf.columns if HPdf[f].dtype == 'O']
for f in categorical_features:
    temp = HPdf.groupby(f)['SalePrice'].count()/len(HPdf)
    temp_df = temp[temp>0.01].index
    HPdf[f] = np.where(HPdf[f].isin(temp_df), HPdf[f],'Rare_var')
```

```python
# converting categorical values to numerical values
from sklearn.preprocessing import LabelEncoder
for f in categorical_features:
    encoder = LabelEncoder()
    HPdf[f] = encoder.fit_transform(HPdf[f])
```

# Feature Scaling

```python
Features_tobe_scaled = [f for f in HPdf.columns if f not in ['Id','SalePrice']]
len(Features_tobe_scaled)
```

```python
# using min max scaler because of more accuracy
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(HPdf[Features_tobe_scaled])
scaler.transform(HPdf[Features_tobe_scaled])
```

```python
# concat Id and SalePrice to the train data
train_data = pd.concat([HPdf[['Id','SalePrice']].reset_index(drop=True), 
                  pd.DataFrame(scaler.transform(HPdf[Features_tobe_scaled]), 
                               columns=Features_tobe_scaled)], axis=1)
```

```python
train_data.head()
```

```python
train_data.to_csv('X_train.csv', index=False)
```

# Feature Selection

```python
y_train = train_data[['SalePrice']]
```

```python
X_train = train_data.drop(['Id','SalePrice'], axis=1)
```

```python
# applying feature selection using Lasso
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))
feature_sel_model.fit(X_train, y_train)
```

```python
# true false says which model to consider and which to not
feature_sel_model.get_support()
```

```python
selected_features = X_train.columns[(feature_sel_model.get_support())]

print(f'The total features: {X_train.shape[1]}')
print(f'selected features: {len(selected_features)}')
```

```python
X_train = X_train[selected_features]
```

```python
X_train.head()
```

```python
X_train.to_csv('X_train.csv', index= False)
```

```python
y_train.to_csv('y_train.csv', index=False)
```

# Doing the same for test data

```python
HPdfT = pd.read_csv("/content/drive/MyDrive/Kaggle/FE_Advance_HousePrice/test.csv")
```

```python
HPdfT.shape
```

```python
# getting numerical for test set
num_features_t = [f for f in HPdfT.columns if HPdfT[f].dtype != 'O']

# getting year features
year_features_t = [f for f in num_features_t if "Yr" in f or "Year" in f]

# changing year features
for f in year_features_t:
    if f != 'YrSold':
        HPdfT[f] = HPdfT['YrSold']-HPdfT[f]

# finding missing values for numerical features
missing_num_features_t = [f for f in num_features_t if HPdfT[f].isnull().sum()>1 ]

# replacing missing values with median values
for f in missing_num_features_t:
    # as we have seen that more outliers in EDA, we replace with the median value
    median_value = HPdfT[f].median()
    HPdfT[f].fillna(median_value, inplace=True)
```

```python
# getting categorical features for test set
categorical_features_t = [f for f in HPdfT.columns if HPdfT[f].dtype == 'O']

# features with missing value
missing_cat_features_t = [f for f in categorical_features_t if HPdfT[f].isnull().sum()>1]

# replacing all the missing values in categorical_features_t function
def replace_cat_feature_t(HPdfT, missing_cat_features_t):
    testdata = HPdfT.copy()
    testdata[missing_cat_features_t] = testdata[missing_cat_features_t].fillna('Missing')
    return testdata

HPdfT = replace_cat_feature_t(HPdfT, missing_cat_features_t)
HPdfT.head()
```

```python
# converting skewed distribution feature to log normal distribution
skewed_features_t = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea']
for f in skewed_features_t:
    HPdfT[f] = np.log(HPdfT[f])
    
```

```python
# converting categorical values to numerical values
from sklearn.preprocessing import LabelEncoder
for f in categorical_features_t:
    encoder = LabelEncoder()
    HPdfT[f] = encoder.fit_transform(HPdfT[f])
```
